```python
# W5a1
def Max_of_Two(a, b):
    if a > b:
        return a
    return b


a, b = map(int, input().split())
print(Max_of_Two(a, b))
```

```python
# W5a2
def Swap(a, b):
    a, b = b, a
    return a, b

a, b = map(int, input().split())
print(Swap(a, b))
```

```python
# W5a3 
import math
def prime_check(n):
    if n < 2:
        return False
    for i in range(2, math.isqrt(n) + 1):
        if n % i == 0:
            return False
    return True

n = int(input())
print(prime_check(n))
```

```python
# W5a4 
import math
def perfect_number_check(n):
    if n < 6:
        return False
    fac_sum = 0    
    for i in range(1, math.isqrt(n) + 1):
        if n % i == 0:
            fac_sum += i
            fac_sum += (n // i)
    return fac_sum - n == n

n = int(input()) 
print(perfect_number_check(n))
```

```python
# W5a5
def find_k_in_list(k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i + 1
    return -1

lst = list(map(int, input().split()))
k = int(input()) 
print(find_k_in_list(k))
```

```python
# W5a6 
def factorial(n):
    if n < 0:
        return 'Invalid' 
    elif n == 0 == 1: 
        return 1
    else:
        ket_qua = 1
        for i in range(2, n + 1):
            ket_qua *= i 
        return ket_qua

n = int(input()) 
print(factorial(n))
```

```python
# W5a7
def calculator(num1, num2, operat):
    if operat == '-':
        return round(num1-num2, 2)
    elif operat == '+':
        return round(num1+num2, 2)
    elif operat == '/':
        return round(num1/num2, 2)
    elif operat == '*':
        return round(num1*num2, 2)
    
n1, n2, operat = input().split()
num1, num2 = int(n1), int(n2)
print(calculator(num1, num2, operat))
```

```python
# W5a8
def hamming_distance(x, y): # OR XOR AND
    xor_result = x ^ y # XOR: 1 + 0 = 1, 0 + 0 = 0, 1 + 1 = 0          
    count = 0          # 1: 001, 4: 100 ==> xor_result = 001
    while xor_result:                                   #001
        count += xor_result & 1 # AND: 1 + 1 = 1
        xor_result >>= 1
    return count

x, y = map(int, input().split())
print(hamming_distance(x, y))
```

```python
# W5a9
def sum_of_each_num(n):
    sum_num = 0
    while n > 0:
        sum_num = sum_num + (n % 10)
        n //= 10
    return sum_num
    
n = int(input()) 
print(sum_of_each_num(n))
```
