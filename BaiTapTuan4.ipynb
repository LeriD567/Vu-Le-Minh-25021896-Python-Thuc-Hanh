```python
## W4A2
while 1:
    n = int(input())
    if n > 0:
        is_prime = True
        if n == 1:
            is_prime = False
        else:
            for i in range(2, int(n**(1/2)+1)):
                if n % i == 0:
                    is_prime = False
                    break

        if is_prime:
            print(f'{n} là số nguyên tố')
        else:
            print(f'{n} không phải số nguyên tố')

        break
    else:
        print('Hãy nhập n dương')
```

```python
#W4a6
import math

def nguyen_to(n):
    if n < 2:
        return False
    for i in range(2, math.isqrt(n) + 1):
        if n % i == 0:
            return False
    return True


while 1:
    try:
        a, b = map(int, input('Nhập vào a, b (a <= b)').split())
        total = 0
        for num in range(a, b+1):
            if nguyen_to(num):
                total += num
        print(total)    
    except:
        break
        
```

```python
#W4a7
import math
def nguyen_to(n):
    if n < 2:
        return False
    for i in range(2, math.isqrt(n) + 1):
        if n % i == 0:
            return False
    return True
    
n = int(input())
ans = n
while ans >= 2:
    if n % ans == 0 and nguyen_to(ans):
        print(ans)
        break
    ans -= 1
```

```python
#W4a8
n = int(input())

def reverse(n):
    v = 0
    n = abs(n)
    while n != 0: # Chia n lấy dư rồi cộng với cái số dư lúc trước đã đc đẩy lên 1 hàng, cắt phần thập phân
        v = v*10 + n%10
        n //= 10
    return v

count = 0
while n != reverse(n):
    n += reverse(n)
    count += 1
    if count > 100:
        break
if count > 100:
    print('Số quá lớn, ko thể tính')
else:
    print(f'Số bước {count}, Số đối xứng {n}')
```

```python
#W4a9
import math
n = int(input())

def ktra_chu_so_khac_nhau(num):
    str_num = str(num)
    return len(str_num) == len(set(str_num)) #Do phần tử trong set là độc nhất

result = []
for i in range(1, math.isqrt(n)+1):
    square = i * i
    if ktra_chu_so_khac_nhau(square):
        result.append(square)

if result:
    print(*result)
else:
    print('(no number)')
```

```python
#W4a10
khoang_n = int(input())
def length_Collatz(n):
    length = 0
    while n != 1:
        if n % 2 == 0:
            n //= 2
        else:
            n = 3*n + 1
        length += 1
    return length

x = 1
max_length = 0
for i in range(1, khoang_n + 1):
    current_length = length_Collatz(i)
    if max_length < current_length:
        max_length = current_length
        x = i

print(x, max_length)
```

```python
#W4a13
so1, so2 = map(int, input().split())
def check_factor(n):
    sum_of_factors = 0
    for i in range(1, n ):
        if n % i == 0:
            sum_of_factors += i 
    return sum_of_factors

if so1 == check_factor(so2) and so2 == check_factor(so1):
    print('true')
else:
    print('false')
```

```python
#W4a14
m, n = map(int, input().split())
u = min(m, n)
while u > 0:
    x = 1
    for i in range(2, min(m, n) + 1):
        if m % i == n % i == 0:
            x = i
    u -= 1
    
print(x)
```

```python
#W4a15
m, n = map(int, input().split())
cho = 4
ga = 2

def tim_con(tong_con, tong_chan):
    for i in range(1, tong_con + 1):
        j = tong_con - i
        if i*cho + j*ga == tong_chan:
            return f'Số chó: {i}, Số gà: {j}'
        else:
            return 'Invalid'

print(tim_con(m, n))
```
