```python
# W7a1
def binary_search(arr, target):
    left, right = 0, (len(arr) - 1) 
    while left <= right: 
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid 
        elif arr[mid] < target: 
            left = mid + 1
        else:
            right = mid - 1
    return -1 

arr = list(map(int, input().split()))
arr.sort() 
target = int(input()) 

print(binary_search(arr, target))
```

```python
# W7a2 
def count_occurrences(arr, x): 
    cnt = 0 
    for i in arr:
        if i == x:
            cnt += 1
    return cnt 

arr = list(map(int, input().split()))
x = int(input()) 

print(count_occurrences(arr, x)) 
```

```python
# W7a3 
def bubble_sort(arr): 
    n = len(arr)
    for i in range(n):
        swapped = False

        for j in range(n - i - 1): 
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
      
        if not swapped:
            break 
    return arr

def selection_sort(arr): 
    n = len(arr)
    for i in range(n): 
        min_index = i 

        for j in range(i+1, n): 
            if arr[j] < arr[min_index]: 
                min_index = j 

        arr[i], arr[min_index] = arr[min_index], arr[i] 
    return arr

arr = list(map(int, input().split())) 

print(bubble_sort(arr)) 
print(selection_sort(arr)) 
```

```python
# W7a4 
def max_occurrences(arr): 
    if not arr:
        return 'Rỗng' 
        
    counts = {}
    max_cnt = 0
    the_num = None 

    for num in arr:
        if num in counts: 
            counts[num] += 1
        else:
            counts[num] = 1 

    for num in counts: 
        if counts[num] > max_cnt:
            the_num = num 
            max_cnt = counts[num] 

    return f'{the_num} xuat hien nhieu nhat, som nhat, {max_cnt} lan' 

arr = list(map(int, input().split())) 

print(max_occurrences(arr)) 
```

```python
# W7a5 
def cnt_sum_pair(arr, X):
    cnt = 0
    
    for i in range(len(arr)):
        for j in range(i+1, len(arr)):
            if arr[i] + arr[j] == X:
                cnt += 1

    return cnt 

arr = list(map(int, input().split()))
X = int(input()) 

print(cnt_sum_pair(arr, X)) 

# improvement: 
# def cnt_sum_pair_optimized(arr, X):
#     cnt = 0
#     seen_numbers = {} # Dictionary lưu các số đã gặp và tần suất của nó

#     for num in arr:
#         target = X - num # Số cần tìm để ghép cặp
        
#         # Nếu target đã xuất hiện trước đó trong danh sách
#         if target in seen_numbers:
#             # Cộng thêm số lần target đã xuất hiện vào biến đếm
#             cnt += seen_numbers[target]
        
#         # Thêm số hiện tại vào danh sách "đã gặp"
#         seen_numbers[num] = seen_numbers.get(num, 0) + 1
            
#     return cnt
```
