```python
# Write code below ðŸ’–
print('Q1) Do you like Dawn or Dusk?\n    1) Dawn\n    2) Dusk')
ans = int(input('Your ans: '))
gry = 0
rav = 0
huf = 0
sly = 0
if ans == 1:
  gry += 1
  rav += 1
elif ans == 2:
  huf += 1
  sly += 1
else:
  print('Wrong input')

print('Q2) When Iâ€™m dead, I want people to remember me as:\n    1) The Good\n    2) The Great\n    3) The Wise\n    4) The Bold')
ans = int(input('Your ans: '))
if ans == 1:
  huf += 2
elif ans == 2:
  sly += 2
elif ans  == 3:
  rav += 2
elif ans == 4:
  gry += 2
else:
  print('Wrong input')

print('Q3) Which kind of instrument most pleases your ear?\n    1) The violin\n   2) The trumpet\n    3) The piano\n   4) The drum')
ans = int(input('Your ans: '))
if ans == 1:
  sly += 4
elif ans == 2:
  huf += 4
elif ans  == 3:
  rav += 4
elif ans == 4:
  gry += 4
else:
  print('Wrong input')

max_score = max(gry, rav, huf, sly)

print('Gryffindor:', gry)
print('Ravenclaw:', rav)
print('Hufflepuff:', huf)
print('Slytherin:', sly)
if max_score == gry:
  print('You are a Gryffindor')
elif max_score == rav:
  print('You are a Ravenclaw')
elif max_score == huf:
  print('You are a Hufflepuff')
elif max_score == sly:
  print('You are a Slytherin')
```
