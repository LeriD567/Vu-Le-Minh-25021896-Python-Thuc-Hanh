```python
# W6a4 
key_value = input().split()  # nhận vào những dữ liệu key_value, dạng list 

def anh_xa_dict(k_v):
    mirror = {}  # tạo sẵn 1 dict rỗng
    
    for item in k_v:  # ktra từng cặp trong key_value
        k, v = item.split(':')  # tách key = k và value = v
        
        if k not in mirror:  # ktra k trong dict mirror 
            mirror[k] = [v]  # nếu k có thì add thêm k với value v(đc tạo dạng list) 
        else:
            mirror[k].append(v)  # nếu có sẵn k thì thêm v vào list v đã đc tạo trc 

    return mirror

print(anh_xa_dict(key_value)) 
```

```python
# W6a5
int_list = list(map(int, input().split())) 

def xet_dau(n): 
    xet_dau_dict = {'positives': 0, 'negatives': 0, 'zeros': 0}
    
    for num in n: 
        if num > 0: 
            xet_dau_dict['positives'] += 1
        elif num < 0: 
            xet_dau_dict['negatives'] += 1
        else:
            xet_dau_dict['zeros'] += 1  
    return xet_dau_dict

print(xet_dau(int_list)) 
```

```python
# W6a9
lenght_dict1 = int(input('Độ dài dict 1: ')) 
lenght_dict2 = int(input('Độ dài dict 2: '))

def nhap_key_value(n): 
    d = {}
    for i in range (0, n):
        key = input('Key: ')
        value = int(input('Value: ')) 
        d[key] = value 
    return d 

dic1 = nhap_key_value(lenght_dict1) 
dic2 = nhap_key_value(lenght_dict2)

dic3 = dic1.copy() 
for key, value in dic2.items():
    if key not in dic3:
        dic3[key] = value
    else:
        dic3[key] += value 

final_dict = dict(sorted(dic3.items(), key = lambda item: item[0]))  # có thể bỏ lambda vì sorted mặc định xét key
        

print(final_dict)
```

```python
# W6a10
int_list = list(map(int, input().split()))
k = int(input()) 

sum_tuple = []

for i in range(len(int_list)):
    for j in range(i+1, len(int_list)):
        if int_list[i] + int_list[j] == k:
            sum_tuple.append((int_list[i], int_list[j])) 

sum_tuple.sort()
            
print(sum_tuple) 
```

```python
# W6a13 
lenght_dict = int(input('Độ dài dict: ')) 

def nhap_key_value(n): 
    d = {}
    for i in range (0, n):
        key = input('Key: ')
        value_str = input('Value: ')

        try:  # thử chuyển thành Int 
            value = int(value_str) 
        except ValueError: 
            try:  # nếu k phải Int thử Float
                value = float(value_str) 
            except ValueError: # k phải Float giữ nguyên Str
                value = value_str
        d[key] = value 
    return d 

org_dict = nhap_key_value(lenght_dict) 
reverse_key_value_dict = {}

for key, value in org_dict.items(): 
    reverse_key_value_dict[value] = key

print(reverse_key_value_dict)
```
