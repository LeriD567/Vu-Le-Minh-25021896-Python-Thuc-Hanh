```python
# W6a16 
n, m = map(int, input().split()) 

matrix = []

for i in range(n): 
    row = list(map(int, input().split())) 
    matrix.append(row) 

# in ma trận có định dạng
for i in range(n):
    for j in range(m): 
        print(f'{matrix[i][j]:4d}', end = '') 
    print() # xuống dòng
```

```python
# W6a19 
int_list = list(map(int, input().split()))

max_value = max(int_list) 
list_max = []

for i in range(len(int_list)): 
    if int_list[i] == max_value: 
        list_max.append(i+1)

print(list_max)
    
```

```python
# W6a20 
int_list = list(map(int, input().split()))
k = int(input()) 

def find_k_in_list(int_list, k): 
    for i in range(len(int_list)): 
        if k not in int_list: 
            return -1
            break
        elif int_list[i] == k:
            return i+1 
            break

print(find_k_in_list(int_list, k))
```

```python
# W6a21 
string = input()
str_space = string.split()

position_dict = {}

for index, value in enumerate(str_space):
    if value not in position_dict: 
        position_dict[value] = (index,) 
    else:
        position_dict[value] = position_dict[value] + (index,)

print(position_dict)
```

```python
# W6a22 
int_list = list(map(int, input().split())) 

def find_max_in_list(int_list):
    max_list = max(int_list) 
    index_list = []
    for i in range(len(int_list)): 
        if int_list[i] == max_list: 
            index_list.append(i)
    return max_list, index_list

print(*find_max_in_list(int_list)) 
```
